## 1. Calculate the Total Bill
A restaurant bill is ₹450. GST is 18%. Calculate the final bill amount.

```python
# Write your solution here
bill=450

gst=0.18*bill
sgst=0.12*bill
other=55
table_charge=55

final_bill=bill+gst+sgst+other+table_charge

print(final_bill)
```
