## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

### Approach:
Use two counters and check with modulo operator.


```python
# write your code here
n = int(input("enter a number here: "))

even = 0
odd = 0

for i in range(1, n+1):
    if i % 2 == 0:   # check each number
        even += 1
        print(i, "is even")
    else:
        odd += 1
        print(i, "is odd")

print("Total even numbers:", even)
print("Total odd numbers:", odd)

```

Take a number as input and check whether it is prime or not.

Loop from 2 to N-1 and use `break` if divisible.


```python
# write your code her
n=int(input("enter a  number here"))
if n<=1:
    print("not a prime number")
else:
    for i in range(2,n):
        if n%i==0:
            print(" not a prime number")
            break
    else:
        print("prime number")

       
```

## Question 20 (Hard)

**Title:** Perfect Number Check

### Problem Statement:
Take a number as input and check whether it is a perfect number or not.

### Approach:
Find sum of all factors using a loop and compare with the number.


```python
# write your code here
n = int(input("Enter a number: "))

sum_factors = 0

for i in range(1, n):      
    if n % i == 0:        
        sum_factors += i

if sum_factors == n:
    print("Perfect number")
else:
    print("Not a perfect number")
```
